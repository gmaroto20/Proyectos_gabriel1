```python
#ahora el objetivo es que sume los numeros al dar al +
#habra que crear una variable (global resultado) que almacene la suma de los valores que sumemos
#la suma se hara a traves de la funcion suma (ojo hasta ahora introduciamos texto, para numeros, int)
#para que aparezcan en pantalla se enviaran con un .set(), en este caso la nueva variable resultado
```

```python
from tkinter import *

raiz=Tk()
raiz.title("Calculadora Gabi")

miframe=Frame()
miframe.pack()

operacion=""

resultado=0

#---------------------- diseño pantalla------------------------

pulsotec=StringVar()


pantalla=Entry(miframe,textvariable=pulsotec)
pantalla.grid(row=3,column=1,padx=10,pady=10,columnspan=4)
pantalla.config(bg="yellow",justify="right")

#-----------------------pulsaciones teclado-------------------------

def boton4(num):
    global operacion
    if operacion!="":
        pulsotec.set(num)
        operacion=""
    else:
        pulsotec.set(pulsotec.get()+num)
    
#--------------------- boton suma--------------------------------------
    
def suma(num):
    global operacion
    
    global resultado
    
    resultado=resultado+int(num)
    pulsotec.set(resultado)
    
    operacion="suma"

#----------------- botones del teclado fila 1 row 7-----------------------

tecla7=Button(miframe,command=lambda:boton4("7"))
tecla7.grid(row=7,column=1)
tecla7.config(text="7",width="3")

tecla8=Button(miframe,command=lambda:boton4("8"))
tecla8.grid(row=7,column=2)
tecla8.config(text="8",width="3")

tecla9=Button(miframe,command=lambda:boton4("9"))
tecla9.grid(row=7,column=3)
tecla9.config(text="9",width="3")

tecladiv=Button(miframe,command=lambda:boton4("/"))
tecladiv.grid(row=7,column=4)
tecladiv.config(text="/",width="3")

#----------------- botones del teclado fila 2 row 8-----------------------


    
tecla4=Button(miframe,command=lambda:boton4("4"))
tecla4.grid(row=8,column=1)
tecla4.config(text="4",width="3")

tecla5=Button(miframe,command=lambda:boton4("5"))
tecla5.grid(row=8,column=2)
tecla5.config(text="5",width="3")

tecla6=Button(miframe,command=lambda:boton4("6"))
tecla6.grid(row=8,column=3)
tecla6.config(text="6",width="3")

teclamul=Button(miframe,command=lambda:boton4("x"))
teclamul.grid(row=8,column=4)
teclamul.config(text="x",width="3")


#----------------- botones del teclado fila 3 row 9-----------------------

tecla1=Button(miframe,command=lambda:boton4("1"))
tecla1.grid(row=9,column=1)
tecla1.config(text="1",width="3")

tecla2=Button(miframe,command=lambda:boton4("2"))
tecla2.grid(row=9,column=2)
tecla2.config(text="2",width="3")

tecla3=Button(miframe,command=lambda:boton4("3"))
tecla3.grid(row=9,column=3)
tecla3.config(text="3",width="3")

teclasum=Button(miframe,command=lambda:suma(pulsotec.get()))
teclasum.grid(row=9,column=4)
teclasum.config(text="+",width="3")

#----------------- botones del teclado fila 4 row 10-----------------------


tecla0=Button(miframe,command=lambda:boton4("0"))
tecla0.grid(row=10,column=1)
tecla0.config(text="0",width="3")

teclacom=Button(miframe)
teclacom.grid(row=10,column=2)
teclacom.config(text=",",width="3")

teclares=Button(miframe)
teclares.grid(row=10,column=3)
teclares.config(text="-",width="3")

teclaigu=Button(miframe)
teclaigu.grid(row=10,column=4)
teclaigu.config(text="=",width="3")

raiz.mainloop()
```

```python
from tkinter import *

raiz=Tk()
raiz.title("Calculadora Gabi")

miframe=Frame()
miframe.pack()

operacion=""

resultado=0

#---------------------- diseño pantalla------------------------

pulsotec=StringVar()


pantalla=Entry(miframe,textvariable=pulsotec)
pantalla.grid(row=3,column=1,padx=10,pady=10,columnspan=4)
pantalla.config(bg="yellow",justify="right")

#-----------------------pulsaciones teclado-------------------------

def boton4(num):
    global operacion
    if operacion!="":
        pulsotec.set(num)
        operacion=""
    else:
        pulsotec.set(pulsotec.get()+num)
    
#--------------------- boton suma--------------------------------------
    
def suma(num):
    global operacion
    
    global resultado
    
    resultado=resultado+int(num)
    pulsotec.set(resultado)
    
    operacion="suma"
    
#----------------------funcion igual-----------------------

def elresultado():
    
    global resultado
    
    pulsotec.set(resultado+int(pulsotec.get()))
    
    resultado=0

#----------------- botones del teclado fila 1 row 7-----------------------

tecla7=Button(miframe,command=lambda:boton4("7"))
tecla7.grid(row=7,column=1)
tecla7.config(text="7",width="3")

tecla8=Button(miframe,command=lambda:boton4("8"))
tecla8.grid(row=7,column=2)
tecla8.config(text="8",width="3")

tecla9=Button(miframe,command=lambda:boton4("9"))
tecla9.grid(row=7,column=3)
tecla9.config(text="9",width="3")

tecladiv=Button(miframe,command=lambda:boton4("/"))
tecladiv.grid(row=7,column=4)
tecladiv.config(text="/",width="3")

#----------------- botones del teclado fila 2 row 8-----------------------


    
tecla4=Button(miframe,command=lambda:boton4("4"))
tecla4.grid(row=8,column=1)
tecla4.config(text="4",width="3")

tecla5=Button(miframe,command=lambda:boton4("5"))
tecla5.grid(row=8,column=2)
tecla5.config(text="5",width="3")

tecla6=Button(miframe,command=lambda:boton4("6"))
tecla6.grid(row=8,column=3)
tecla6.config(text="6",width="3")

teclamul=Button(miframe,command=lambda:boton4("x"))
teclamul.grid(row=8,column=4)
teclamul.config(text="x",width="3")


#----------------- botones del teclado fila 3 row 9-----------------------

tecla1=Button(miframe,command=lambda:boton4("1"))
tecla1.grid(row=9,column=1)
tecla1.config(text="1",width="3")

tecla2=Button(miframe,command=lambda:boton4("2"))
tecla2.grid(row=9,column=2)
tecla2.config(text="2",width="3")

tecla3=Button(miframe,command=lambda:boton4("3"))
tecla3.grid(row=9,column=3)
tecla3.config(text="3",width="3")

teclasum=Button(miframe,command=lambda:suma(pulsotec.get()))
teclasum.grid(row=9,column=4)
teclasum.config(text="+",width="3")

#----------------- botones del teclado fila 4 row 10-----------------------


tecla0=Button(miframe,command=lambda:boton4("0"))
tecla0.grid(row=10,column=1)
tecla0.config(text="0",width="3")

teclacom=Button(miframe)
teclacom.grid(row=10,column=2)
teclacom.config(text=",",width="3")

teclares=Button(miframe)
teclares.grid(row=10,column=3)
teclares.config(text="-",width="3")

teclaigu=Button(miframe,command=lambda:elresultado())
teclaigu.grid(row=10,column=4)
teclaigu.config(text="=",width="3")

raiz.mainloop()
```
