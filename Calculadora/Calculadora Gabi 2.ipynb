```python
#vamos a dar utilidad a los botones (que al marcarlos ejecuten la orden encomendada)
#para ello utilizamos una variable de control(StringVar,IntVAr,DoubleVar,BoolVar) pepe=StringVar()
#estas variable almacenan segun sea texto,numero,decimal o booleano el valor asignado a un Widget
#se incluyen en el widget(pantalla por ej) mediante el argumento textvariable=pepe
#para introducir su valor se hace a traves de un metodo pepe.set("valor") que se crea en
#una funcion(tecleoboton) def tecleoboton():
#si se introducen al pulsar un boton(command) por ej. habra que configurar en este la propiedad
#command=tecleoboton
#es decir, en realidad estamos relacionando dos widgets (boton y pantalla)
#al pulsar un boton aparece un numero por pantalla
```

```python
from tkinter import *

raiz=Tk()
raiz.title("Calculadora Gabi")

miframe=Frame()
miframe.pack()

#---------------------- diseño pantalla------------------------

pulsotec=StringVar()


pantalla=Entry(miframe,textvariable=pulsotec)
pantalla.grid(row=3,column=1,padx=10,pady=10,columnspan=4)
pantalla.config(bg="yellow",justify="right")

#-----------------------pulsaciones teclado-------------------------

def boton4():
    pulsotec.set("4")

#----------------- botones del teclado fila 1 row 7-----------------------

tecla7=Button(miframe)
tecla7.grid(row=7,column=1)
tecla7.config(text="7",width="3")

tecla8=Button(miframe)
tecla8.grid(row=7,column=2)
tecla8.config(text="8",width="3")

tecla9=Button(miframe)
tecla9.grid(row=7,column=3)
tecla9.config(text="9",width="3")

tecladiv=Button(miframe)
tecladiv.grid(row=7,column=4)
tecladiv.config(text="/",width="3")

#----------------- botones del teclado fila 2 row 8-----------------------


    
tecla4=Button(miframe,command=boton4)
tecla4.grid(row=8,column=1)
tecla4.config(text="4",width="3")

tecla5=Button(miframe)
tecla5.grid(row=8,column=2)
tecla5.config(text="5",width="3")

tecla6=Button(miframe)
tecla6.grid(row=8,column=3)
tecla6.config(text="6",width="3")

teclamul=Button(miframe)
teclamul.grid(row=8,column=4)
teclamul.config(text="x",width="3")


#----------------- botones del teclado fila 3 row 9-----------------------

tecla1=Button(miframe)
tecla1.grid(row=9,column=1)
tecla1.config(text="1",width="3")

tecla2=Button(miframe)
tecla2.grid(row=9,column=2)
tecla2.config(text="2",width="3")

tecla3=Button(miframe)
tecla3.grid(row=9,column=3)
tecla3.config(text="3",width="3")

teclasum=Button(miframe)
teclasum.grid(row=9,column=4)
teclasum.config(text="+",width="3")

#----------------- botones del teclado fila 4 row 10-----------------------


tecla0=Button(miframe)
tecla0.grid(row=10,column=1)
tecla0.config(text="0",width="3")

teclacom=Button(miframe)
teclacom.grid(row=10,column=2)
teclacom.config(text=",",width="3")

teclares=Button(miframe)
teclares.grid(row=10,column=3)
teclares.config(text="-",width="3")

teclaigu=Button(miframe)
teclaigu.grid(row=10,column=4)
teclaigu.config(text="=",width="3")

raiz.mainloop()
```

```python
from tkinter import *

raiz=Tk()
raiz.title("Calculadora Gabi")

miframe=Frame()
miframe.pack()

#---------------------- diseño pantalla------------------------

pulsotec=StringVar()


pantalla=Entry(miframe,textvariable=pulsotec)
pantalla.grid(row=3,column=1,padx=10,pady=10,columnspan=4)
pantalla.config(bg="yellow",justify="right")

#-----------------------pulsaciones teclado-------------------------

def boton4():
    pulsotec.set(pulsotec.get()+"4")

#----------------- botones del teclado fila 1 row 7-----------------------

tecla7=Button(miframe)
tecla7.grid(row=7,column=1)
tecla7.config(text="7",width="3")

tecla8=Button(miframe)
tecla8.grid(row=7,column=2)
tecla8.config(text="8",width="3")

tecla9=Button(miframe)
tecla9.grid(row=7,column=3)
tecla9.config(text="9",width="3")

tecladiv=Button(miframe)
tecladiv.grid(row=7,column=4)
tecladiv.config(text="/",width="3")

#----------------- botones del teclado fila 2 row 8-----------------------


    
tecla4=Button(miframe,command=boton4)
tecla4.grid(row=8,column=1)
tecla4.config(text="4",width="3")

tecla5=Button(miframe)
tecla5.grid(row=8,column=2)
tecla5.config(text="5",width="3")

tecla6=Button(miframe)
tecla6.grid(row=8,column=3)
tecla6.config(text="6",width="3")

teclamul=Button(miframe)
teclamul.grid(row=8,column=4)
teclamul.config(text="x",width="3")


#----------------- botones del teclado fila 3 row 9-----------------------

tecla1=Button(miframe)
tecla1.grid(row=9,column=1)
tecla1.config(text="1",width="3")

tecla2=Button(miframe)
tecla2.grid(row=9,column=2)
tecla2.config(text="2",width="3")

tecla3=Button(miframe)
tecla3.grid(row=9,column=3)
tecla3.config(text="3",width="3")

teclasum=Button(miframe)
teclasum.grid(row=9,column=4)
teclasum.config(text="+",width="3")

#----------------- botones del teclado fila 4 row 10-----------------------


tecla0=Button(miframe)
tecla0.grid(row=10,column=1)
tecla0.config(text="0",width="3")

teclacom=Button(miframe)
teclacom.grid(row=10,column=2)
teclacom.config(text=",",width="3")

teclares=Button(miframe)
teclares.grid(row=10,column=3)
teclares.config(text="-",width="3")

teclaigu=Button(miframe)
teclaigu.grid(row=10,column=4)
teclaigu.config(text="=",width="3")

raiz.mainloop()
```
